# <p style="text-align: center">PLOTTING</p>

### PLOT TYPES
- plot(kind='line') => (Best for tracking changes over a period of time).
- plot(kind='bar') => (Best for tracking changes over a period of time and for comparing between different group).
- plot(kind='barh') => (Same as bar graph but with an axis change).
- plot(kind='pie') => (Best for comparing parts of whole).



```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

## Plot Types Examples

### What were the different sports in the first olympics ? Plot them using different graphs.

```python
fo = df[df.Edition == 1896]
fo.head()
```

### Line Plot

```python
fo.Sport.value_counts().plot(kind='line');
```

### Bar Plot

```python
fo.Sport.value_counts().plot(kind='bar');
```

### Horizontal Bar Plot with Color Specified

```python
fo.Sport.value_counts().plot(kind='barh', color='green');
```

### Pie Plot

```python
fo.Sport.value_counts().plot(kind='pie');
```

### figsize() => Size of the graph

```python
fo.Sport.value_counts().plot(figsize=(12,3));
```

### Colormaps
- Classes of colormaps
    - Sequential: Used for information that has order.
    - Diverging: Used when the info deviates arount a middle value.
    - Qualitative: Used for information that doesn't has order or relationship.


```python
fo.Sport.value_counts().plot(kind='pie', colormap='Pastel1');
```
